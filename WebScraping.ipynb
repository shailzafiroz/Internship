# 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,company name, CTC, and apply date.

```python
page =requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
jtitle=soup.find_all("div",class_="heading_4_5 profile")
jtitle
```

```python
jobtitle=[]
for i in jtitle:
    for j in i.find_all('a'):
        jobtitle.append(j.text.replace("\n"," "))
    
jobtitle
```

```python
c_name=soup.find_all("div",class_="heading_6 company_name")
c_name
companyname=[]
for i in c_name:
    for j in i.find_all('a'):
        companyname.append(j.text.replace("\n"," "))
    
companyname
```

```python
c=soup.find_all("div",class_="item_body")
c
ct=[]
for i in c:
    ct.append(i.text.replace("\nStarts\xa0Immediately\n                       ",""))
ct
```

```python
asp=[]
for i in list(ct):
    asp.append(i.replace("\n",""))
    
asp
```

```python
while(" " in asp) :
    asp.remove(" ")
asp
```

```python
applydate=[]
for i in range(1,len(asp),2):
    applydate.append(asp[i])
applydate
```

```python
ctc=[]
for i in range(0,len(asp),2):
    ctc.append(asp[i])
ctc
```

```python
fresherjobs=pd.DataFrame(())
fresherjobs['Job_Title']=jobtitle
fresherjobs['Company_name']=companyname
fresherjobs['CTC']=ctc
fresherjobs['Apply_Date']=applydate
fresherjobs
```
